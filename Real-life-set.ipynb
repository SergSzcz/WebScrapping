```python
import requests, pandas 
from bs4 import BeautifulSoup

l=[]
page = requests.get("https://www.krn.pl/mieszkania,wynajem,malopolskie,1,5,s1?se_type=1&fotol=1&ami=32455&nreg2=271&nreg3=1496&lok=Krak%C3%B3w%2C&sort=5_2&limit=24")
c = page.content

soup = BeautifulSoup(c,'html.parser')

all = soup.find_all("div",{"class":"search-results-list"})
    
for item in all:
    adres = item.find_all('span',{"class","address"})

    powi = item.find_all('strong')
    for address in adres:
        for i,j in enumerate(address.text):
            if j == ',':
                print(address.text[0:i])
                print(address.text[i+2:-1])
                print(" ")
    for powierzchnia in powi:
        print(powierzchnia.text)
```

```python
coll = soup.find_all("a",{"class":"col-sm-12"})
for item in coll:
    address = item.find_all('span',{"class","address"})[0].text
    for i,j in enumerate(address):
        if j == ',':
            print(address[0:i])
            print(address[i+2:-1])
               
    print(item.find("strong").text)

    print(item.find("span",{"class":"price"}).text)
            
    print(' ') 
   
```
